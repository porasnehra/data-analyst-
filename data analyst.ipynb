```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Sample DataFrame
std_api = pd.DataFrame({
    'arrival_date_year': [2015, 2016, None],
    'adults': [2, 1, None],
    'babies': [0, 1, None],
    'total_of_special_requests': [1, None, 0]
})

# Apply imputer only on numeric columns
numeric_cols = std_api.select_dtypes(include=['float64', 'int64']).columns
imputer = SimpleImputer(strategy='mean')
std_api[numeric_cols] = imputer.fit_transform(std_api[numeric_cols])
```

```python
dataset["FirstName"].replace("JON","poras",inplace=True)
dataset.head(4)
```

```python
dataset["BirthDate"]=dataset["BirthDate"].astype("object")
dataset.info()
```

```python
q1=dataset["CustomerKey"].quantile(0.25)
q3=dataset["CustomerKey"].quantile(0.75)
iqr=q3-q1
min_r=q1-(1.5*iqr)
max_r=q3+(1.5*iqr)
min_r,max_r
```

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
ft=FunctionTransformer(funct=np.log1p)
ft.fit(dataset[['CustomerKey']])
dataset["CustomerKey_tf"]=ft.transform(dataset[['CustomerKey']])
sns.distplot(dataset["CustomerKey_tf"])

```
