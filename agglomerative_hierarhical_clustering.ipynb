```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate synthetic dataset for Agglomerative Hierarchical Clustering
# We'll create 4 clusters with 150 samples total, 2 features
X, y_true = make_blobs(n_samples=150, centers=4, cluster_std=0.60, random_state=0)

# Create a DataFrame
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
df['True_Cluster'] = y_true

# Display the first few rows
print(df.head())

# Visualize the dataset
plt.scatter(X[:, 0], X[:, 1], c=y_true, s=50, cmap='viridis')
plt.title('Synthetic Dataset for Clustering')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()

# Save the dataset to CSV for practice
df.to_csv('agglomerative_clustering_dataset.csv', index=False)
print("Dataset saved as 'agglomerative_clustering_dataset.csv'")
```

```python
import scipy.cluster.hierarchy as sc

# plt.figure(figsize=(20,2()))
sc.dendrogram(sc.linkage(df,method='single', metric='euclidean'))
plt.savefig("demo.jpg")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(n_clusters=4,linkage='single')
df["Predict"] = ac.fit_predict(df)

df.head(2)
```

```python
# Check the number of clusters
print("Number of true clusters in the dataset:", len(df['True_Cluster'].unique()))
print("Number of predicted clusters:", len(df['Predict'].unique()))
```
