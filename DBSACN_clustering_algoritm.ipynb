```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons, make_blobs
import matplotlib.pyplot as plt

# Generate synthetic data for DBSCAN practice
# Create two moon-shaped clusters
X_moons, _ = make_moons(n_samples=200, noise=0.1, random_state=42)

# Create two blob clusters with different densities
X_blobs1, _ = make_blobs(n_samples=100, centers=1, cluster_std=0.3, center_box=(-1, 1), random_state=42)
X_blobs2, _ = make_blobs(n_samples=100, centers=1, cluster_std=0.8, center_box=(2, 4), random_state=43)

# Combine the datasets
X = np.vstack([X_moons, X_blobs1, X_blobs2])

# Add some random noise points
noise = np.random.uniform(low=-2, high=4, size=(50, 2))
X = np.vstack([X, noise])

# Create DataFrame
df = pd.DataFrame(X, columns=['feature1', 'feature2'])

# Save to CSV
df.to_csv('dbscan_practice_dataset.csv', index=False)

# Visualize the dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, alpha=0.7)
plt.title('DBSCAN Practice Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()

print("Dataset saved as 'dbscan_practice_dataset.csv'")
print(f"Dataset shape: {df.shape}")
print(df.head())
```

```python
print("\nThis dataset contains two moon-shaped clusters, two blob clusters with different densities, and random noise points for DBSCAN clustering practice.")
```

```python
from sklearn.cluster import DBSCAN
import seaborn as sns

db = DBSCAN(eps=0.2,min_samples=5)
df["Predict"] = db.fit_predict(df)

sns.scatterplot(x="feature1",y="feature2",data=df, hue="Predict")
plt. show()
```
