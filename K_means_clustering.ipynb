```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import seaborn as sns

# Generate sample data for K-means clustering
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
X = pd.DataFrame(X, columns=['feature1', 'feature2'])

# Plot the data
plt.scatter(X['feature1'], X['feature2'], s=50)
plt.title('Random Data for K-means Clustering')
plt.show()
```

```python
from sklearn.cluster import KMeans
# Apply K-means clustering
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
# Plot the clustered data
plt.scatter(X['feature1'], X['feature2'], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-means Clustering Results')
plt.show()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss,marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.grid(axis='both')
plt.show()
```

```python
knn = KMeans(n_clusters=4, random_state=0)
X["fit_predict"] = knn.fit_predict(X)
```

```python
import seaborn as sns
sns.pairplot(X, hue="fit_predict", palette="viridis")
plt.title('K-means Clustering with fit_predict')
plt.show()
```
